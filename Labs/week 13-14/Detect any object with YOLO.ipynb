```python
import cv2
import torch
from ultralytics import YOLO

# Load YOLOv8 model trained for object detection
model = YOLO("yolov8n.pt")  # Using the nano version for speed

# List of class IDs to detect (Simply add new IDs here)
detect_classes = [0, 2, 5, 67, 11, 64, 56]  # 0 = Person, 2 = Car, 5 = Bus

# Open video file or webcam
video_path = 0  # Change to 0 for webcam
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLOv8 detection
    results = model(frame)
    people_count = 0

    for result in results:
        for box in result.boxes:
            class_id = int(box.cls)  # Get detected object class
            
            if class_id in detect_classes:  # Detect only selected classes
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                class_name = model.names[class_id]  # Get the object name
                
                # Count people separately
                if class_id == 0:
                    people_count += 1

                # Draw bounding box
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(frame, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the count on frame
    cv2.putText(frame, f"People Count: {people_count}", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Show the frame
    cv2.imshow("Object Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
