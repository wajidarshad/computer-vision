# HOG

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# Load the image
image_path = "car.png"  # Change this to your image path
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Compute HOG descriptor
hog_features, hog_image = hog(
    gray_image,
    orientations=12,
    pixels_per_cell=(6, 6),
    cells_per_block=(2, 2),
    block_norm="L2-Hys",
    visualize=True,
    transform_sqrt=True
)

# Enhance HOG image contrast
hog_image = exposure.rescale_intensity(hog_image, in_range=(0, 10))

#  Display original and HOG images side by side
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.imshow(image)  # Show original image in color
plt.axis("off")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(hog_image, cmap="gray")  # HOG descriptor visualization
plt.axis("off")
plt.title("Enhanced HOG Descriptor")

plt.show()

# Print feature vector size
print(f"HOG Feature Vector Size: {len(hog_features)}")
```
