## Write Text On Image

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">`cv2.putText()` - Explanation</strong>
    <p style="color: #cccccc;">
        The <code>cv2.putText()</code> function in OpenCV is used to draw text on an image. This can be useful when you want to annotate images with labels, titles, or other types of text-based information. The function allows you to control several aspects of the text, including its position, font, size, color, and thickness.
    </p></br>
    <strong style="color: #ffcc00;">How it works:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Image:</strong> You specify the image where the text will be added.</li>
        <li><strong>Text:</strong> The actual string of text you want to display on the image.</li>
        <li><strong>Position:</strong> The text is placed at a specific location on the image, specified by the <code>org</code> parameter. The position is given as a tuple <code>(x, y)</code>, where <code>x</code> is the horizontal position and <code>y</code> is the vertical position.</li>
        <li><strong>Font:</strong> You can choose from various fonts (e.g., <code>cv2.FONT_HERSHEY_SIMPLEX</code>), with different styles for the text.</li>
        <li><strong>Font Size:</strong> The size of the text is adjusted with the <code>fontScale</code> parameter. The higher the value, the larger the text.</li>
        <li><strong>Text Color:</strong> The color is specified in BGR (Blue, Green, Red) format, which is OpenCV's default color space.</li>
        <li><strong>Thickness:</strong> The thickness parameter allows you to adjust the thickness of the text's outline. Thicker lines make the text more prominent.</li>
        <li><strong>Line Type:</strong> You can select the type of line used to draw the text, with <code>cv2.LINE_AA</code> providing a smoother appearance due to anti-aliasing.</li>
        <li><strong>Text Origin:</strong> The <code>bottomLeftOrigin</code> parameter specifies whether the origin (starting point) of the text is at the bottom-left or top-left corner. By default, OpenCV uses the top-left corner.</li>
    </ul>
    <strong style="color: #ffcc00;">Common Use Cases:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Annotations:</strong> Adding labels or descriptions to images, such as adding the name of objects or landmarks.</li>
        <li><strong>Watermarking:</strong> Placing a watermark or text label over an image to indicate ownership or copyright.</li>
        <li><strong>Timestamps or Date/Time:</strong> Adding time stamps or date annotations to images, especially for photos or video frames.</li>
        <li><strong>Graphical User Interfaces (GUIs):</strong> Displaying dynamic text or messages in OpenCV-based GUI applications.</li>
    </ul>

#### Function to write text at the middle of an Image

```python
def writeText(img, text, color=(120, 255, 80), font = cv2.FONT_HERSHEY_SIMPLEX, font_scale = 1, thickness = 3):
    # Get the image dimensions
    h, w = img.shape[:2]
    
    # Calculate the size of the text
    (text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, thickness)
    
    # Calculate the center position
    center_x = (w - text_width) // 2
    center_y = (h + text_height) // 2  # Add height to vertical center so the text is not cut off
    
    # Put the text on the image
    img = cv2.putText(img, text, (center_x, center_y), font, font_scale, color, thickness, cv2.LINE_AA)

    # Return Image
    return img
```

#### `cv2.HoughLines()` - Explanation

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">`cv2.HoughLines()` - Hough Line Transform</strong>
    <p style="color: #cccccc;">
        The <code>cv2.HoughLines()</code> function in OpenCV is used to detect straight lines in an image using the Hough Line Transform algorithm. This function is useful for detecting lines in images, especially after edge detection has been applied. It works by transforming the image space into a parameter space and identifying the lines that can be represented in this transformed space.
    </p>
    <strong style="color: #ffcc00;">How it works:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Image:</strong> The input image, typically a single-channel (grayscale) image, where edges have been detected (often using the Canny edge detector).</li>
        <li><strong>Rho:</strong> The resolution of the accumulator in pixels. It defines the distance resolution of the accumulator in the Hough space.</li>
        <li><strong>Theta:</strong> The angle resolution of the accumulator in radians. This typically corresponds to <code>np.pi / 180</code>, which is equivalent to 1 degree in radians.</li>
        <li><strong>Threshold:</strong> The minimum number of intersections in the Hough space required to identify a line. A higher value results in fewer lines being detected, while a lower value results in more lines being detected.</li>
    </ul>
    <strong style="color: #ffcc00;">Parameters:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>image:</strong> The input image. This should be a grayscale image with detected edges, usually after applying an edge detection algorithm like the Canny edge detector.</li>
        <li><strong>rho:</strong> The resolution of the accumulator in the Hough space in pixels. This is typically set to 1.</li>
        <li><strong>theta:</strong> The angle resolution of the accumulator in radians. This is typically set to <code>np.pi / 180</code> for 1-degree resolution.</li>
        <li><strong>threshold:</strong> The minimum number of votes (intersections) needed to detect a line. If a line accumulates fewer votes, it won't be considered.</li>
    </ul>
    <strong style="color: #ffcc00;">Returns:</strong>
    <p style="color: #cccccc;">
        The function returns a list of detected lines. Each line is represented by a pair of values <code>[rho, theta]</code>, where:
        <ul>
            <li><strong>rho:</strong> The distance from the origin to the closest point on the line.</li>
            <li><strong>theta:</strong> The angle of the line in radians.</li>
        </ul>
        If no lines are detected, the function returns <code>None</code>.
    </p>
    <strong style="color: #ffcc00;">Explanation of Parameters:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Rho ():</strong> This parameter defines the distance resolution in the Hough space. It represents the distance from the origin to the closest point on the detected line, measured in pixels.</li>
        <li><strong>Theta ():</strong> The angle of the line in polar coordinates, where 0 corresponds to the x-axis. Increasing values of <code>theta</code> represent counterclockwise rotations of the line.</li>
        <li><strong>Threshold:</strong> The threshold defines the minimum number of votes required for a line to be detected. If a line has fewer votes than this threshold, it will not be included in the results.</li>
    </ul>
    <strong style="color: #ffcc00;">Common Use Cases:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Lane Detection in Road Images:</strong> Detecting lanes in images from autonomous vehicles or road monitoring systems.</li>
        <li><strong>Architectural and Engineering Drawings:</strong> Extracting straight lines from technical drawings or blueprints.</li>
        <li><strong>Object Boundary Detection:</strong> Identifying edges of objects in scanned images or photographs.</li>
        <li><strong>Robotic Path Finding:</strong> Detecting paths or walls for robots in indoor environments.</li>
    </ul>
    <strong style="color: #ffcc00;">Important Notes:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li>The <code>cv2.HoughLines()</code> function works best when edges are detected using an algorithm like Canny, as it operates in the space of detected edges.</li>
        <li>The accuracy of line detection depends on the quality of edge detection and the value of the threshold parameter.</li>
        <li>If more precision is needed for line detection, consider using <code>cv2.HoughLinesP()</code>, a probabilistic version that directly provides line endpoints.</li>
    </ul>
    <strong style="color: #ffcc00;">See Also:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><a href="https://docs.opencv.org/4.x/d9/d61/tutorial_hough_lines.html" style="color: #ffcc00;">`cv2.HoughLinesP()`</a>: A probabilistic version of the Hough Line Transform that directly returns endpoints of lines.</li>
        <li><a href="https://docs.opencv.org/4.x/da/d22/tutorial_py_canny.html" style="color: #ffcc00;">`cv2.Canny()`</a>: A function used to perform edge detection, often used as a preprocessing step before applying Hough Line Transform.</li>
    </ul>
</div>

```python
# Load the image
img = r.load_image('linesDetected.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Edge detection
edge = cv2.Canny(gray, 200, 255)

# Hough Line detection
lines = cv2.HoughLines(edge, 1, np.pi/180, 200)  # Use np.pi/180 for theta in radians

# Check if any lines were detected
if lines is not None:
    for line in lines:
        rho, theta = line[0]  # Get rho and theta values
        
        # Calculate the coordinates for the line endpoints
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))  # Start point of the line
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))  # End point of the line
        y2 = int(y0 - 1000 * (a))

        # Draw the line on the image
        cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 11)

# Show the image with lines drawn
r.show_image_with_color(img)
```

#### `cv2.circle()` - Explanation

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">`cv2.circle()` - Drawing Circles</strong>
    <p style="color: #cccccc;">
        The <code>cv2.circle()</code> function in OpenCV is used to draw a circle on an image. This is commonly used for marking specific points, creating circular markers, or highlighting areas within an image. The function allows you to control various aspects of the circle, including its center, radius, color, and thickness.
    </p>
    <strong style="color: #ffcc00;">How it works:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Image:</strong> You specify the image where the circle will be drawn.</li>
        <li><strong>Center:</strong> The <code>center</code> parameter specifies the center of the circle as a tuple <code>(x, y)</code>, where <code>x</code> is the horizontal position and <code>y</code> is the vertical position.</li>
        <li><strong>Radius:</strong> The <code>radius</code> parameter specifies the radius of the circle.</li>
        <li><strong>Color:</strong> The color of the circle is defined in BGR (Blue, Green, Red) format. For example, <code>(255, 0, 0)</code> would be blue, and <code>(0, 255, 0)</code> would be green.</li>

        <li><strong>Shift:</strong> The <code>shift</code> parameter is used to specify the precision of the circle's center and radius. This is typically not necessary unless dealing with high precision values.</li>

#### `cv2.polylines()` - Explanation

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">`cv2.polylines()` - Drawing Polylines or Polygons</strong>
    <p style="color: #cccccc;">
        The <code>cv2.polylines()</code> function in OpenCV is used to draw one or more polygons (or polyline paths) on an image. This is particularly useful for creating shapes, drawing boundaries, or connecting a series of points to represent objects or paths. The function allows you to control various aspects of the polyline, such as its color, thickness, and whether it is closed or open.
    </p>
    <strong style="color: #ffcc00;">How it works:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Image:</strong> You specify the image where the polyline will be drawn.</li>
        <li><strong>Points:</strong> The <code>pts</code> parameter specifies the points that define the polyline. These points should be provided as a list of points, where each point is a tuple <code>(x, y)</code>. If you're drawing multiple polylines, this will be a list of lists.</li>
        <li><strong>Is Closed:</strong> The <code>isClosed</code> parameter is a boolean that indicates whether the polyline should be closed (i.e., the last point is connected to the first point to form a polygon). By default, this is set to <code>False</code>, meaning the polyline is open.</li>
        <li><strong>Color:</strong> The color of the polyline is specified in BGR (Blue, Green, Red) format. For example, <code>(255, 0, 0)</code> would be blue, and <code>(0, 255, 0)</code> would be green.</li>
        <li><strong>Thickness:</strong> The <code>thickness</code> parameter controls the width of the polyline. A higher value will make the polyline thicker.</li>
        <li><strong>Line Type:</strong> You can specify the line type used for the polyline, with <code>cv2.LINE_8</code> being the default. <code>cv2.LINE_AA</code> can be used for anti-aliased lines, which results in smoother edges.</li>
        <li><strong>Shift:</strong> The <code>shift</code> parameter is used to specify the precision of the points' coordinates. This is typically used for higher precision coordinates.</li>
    </ul>
    <strong style="color: #ffcc00;">Common Use Cases:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Drawing Geometric Shapes:</strong> Used to draw polygons (triangles, rectangles, etc.) by defining multiple points.</li>
        <li><strong>Path Visualization:</strong> Used for visualizing paths or trajectories in image processing tasks, such as tracking or movement analysis.</li>
        <li><strong>Object Boundaries:</strong> Useful for marking object boundaries or regions of interest in an image, such as in segmentation tasks or object detection.</li>
        <li><strong>Annotations:</strong> Ideal for annotating complex shapes or areas in an image, such as highlighting specific features or regions.</li>

    <ol style="color: #cccccc; padding-left: 20px;">
        <li>Load an image using <code>cv2.imread()</code>.</li>

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">1. Defining the Points for the Polygon (Triangle)</strong>
    <p style="color: #cccccc;">
        We define the coordinates of the triangles vertices using a NumPy array:
    </p>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>(105, 230):</strong> bottom-left vertex.</li>
        <li><strong>(170, 50):</strong> top vertex.</li>
        <li><strong>(230, 230):</strong> bottom-right vertex.</li>
    </ul>
    <p style="color: #cccccc;">
        The array uses <code>np.int32</code> to store the coordinates as integers, ensuring accurate pixel positioning.
    </p>
    <strong style="color: #ffcc00;">2. Reshaping the Points</strong>
    <p style="color: #cccccc;">
        The <code>reshape((-1, 1, 2))</code> function reshapes the array to match OpenCVs required format:
    </p>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>-1:</strong> Auto-determines the number of rows (3 in this case).</li>
        <li><strong>1:</strong> Represents each point as a single <code>(x, y)</code> pair.</li>
        <li><strong>2:</strong> Represents the x and y coordinates for each point.</li>
    </ul>
    <strong style="color: #ffcc00;">3. Why Reshaping is Necessary</strong>
    <p style="color: #cccccc;">
        OpenCVs <code>cv2.polylines()</code> function expects the points to be in a specific format: a list of 2D arrays containing <code>(x, y)</code> pairs.
    </p>
    <p style="color: #cccccc;">
        The reshaping ensures that the points are in the correct structure for OpenCV to render the polygon properly.
    </p>
</div>


```python
img = r.load_image()
# Define points of a triangle
points = np.array([[105, 230], [170, 50], [230, 230]], np.int32)
points = points.reshape((-1, 1, 2))

# Drawing a polygon (triangle)
img = cv2.polylines(img, [points], isClosed=True, color=(0, 255, 0), thickness=3)

# Show the image with the polygon
r.show_image_with_color(img, 'Image with Polygon')
```

#### `cv2.ellipse()` - Explanation

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: left; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 5px solid #ffcc00; border-radius: 15px; box-shadow: 0 6px 20px rgba(0, 0, 0, 0.6); margin-bottom: 25px;">
    <strong style="color: #ffcc00;">The <code>cv2.ellipse()</code> Function in OpenCV</strong>
    <p style="color: #cccccc;">
        The <code>cv2.ellipse()</code> function in OpenCV is used to draw an ellipse on an image. An ellipse is a geometric shape that looks like a stretched circle and can be used in various applications such as marking elliptical regions of interest or creating oval shapes for visual emphasis. The function allows you to control several aspects of the ellipse, including its center, axes, angle, color, thickness, and more.
    </p>
    <strong style="color: #ffcc00;">How it Works:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Image:</strong> You specify the image where the ellipse will be drawn.</li>
        <li><strong>Center:</strong> The <code>center</code> parameter specifies the center of the ellipse as a tuple <code>(x, y)</code>, where <code>x</code> is the horizontal position and <code>y</code> is the vertical position.</li>
        <li><strong>Axes:</strong> The <code>axes</code> parameter specifies the lengths of the two axes of the ellipse. This is a tuple <code>(majorAxisLength, minorAxisLength)</code>, where <code>majorAxisLength</code> is the length of the major axis and <code>minorAxisLength</code> is the length of the minor axis.</li>
        <li><strong>Angle:</strong> The <code>angle</code> parameter specifies the rotation angle of the ellipse in degrees. A value of <code>0</code> means no rotation.</li>
        <li><strong>Start Angle:</strong> The <code>startAngle</code> parameter specifies the starting angle (in degrees) for the ellipse arc. Typically, this is <code>0</code> to start the arc from the "3 o'clock" position.</li>
        <li><strong>End Angle:</strong> The <code>endAngle</code> parameter specifies the ending angle (in degrees) for the ellipse arc. A full ellipse has an end angle of <code>360</code>.</li>
        <li><strong>Color:</strong> The color of the ellipse is defined in BGR (Blue, Green, Red) format. For example, <code>(0, 255, 0)</code> would be green, and <code>(0, 0, 255)</code> would be red.</li>
        <li><strong>Thickness:</strong> The <code>thickness</code> parameter controls the width of the ellipse's outline. If you set this to <code>-1</code>, the ellipse will be filled with the given color.</li>
        <li><strong>Line Type:</strong> You can choose the type of line used for drawing the ellipse. The default is <code>cv2.LINE_8</code>, but <code>cv2.LINE_AA</code> (anti-aliased) provides smoother results.</li>
        <li><strong>Shift:</strong> The <code>shift</code> parameter specifies the precision of the ellipse's center and axes. This is typically not needed unless you're working with high precision coordinates.</li>
    </ul>
    <strong style="color: #ffcc00;">Common Use Cases:</strong>
    <ul style="color: #cccccc; padding-left: 20px;">
        <li><strong>Ellipse Annotation:</strong> Used for highlighting elliptical regions of interest in an image, such as marking or emphasizing specific areas.</li>
        <li><strong>Object Detection:</strong> Useful in computer vision tasks where objects have an elliptical shape, such as in shape recognition or detection.</li>
        <li><strong>Visualizations:</strong> Commonly used in visualizations or diagrams where an oval shape is needed, such as in graphical overlays.</li>
        <li><strong>Tracking:</strong> Can be used to represent motion paths or tracks that follow an elliptical trajectory.</li>
    </ul>
    <strong style="color: #ffcc00;">Example Workflow:</strong>
    <ol style="color: #cccccc; padding-left: 20px;">
        <li><strong>Load an Image:</strong> Use <code>cv2.imread()</code> to load the image where you want to draw the ellipse.</li>
